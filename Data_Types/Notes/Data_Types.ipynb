## Datatype in Python

-  `int`    (both positive and negative whole numbers, eg : -5, -1, 100, 5)
-  `float`  (both positive and negative decimal numbers, eg : 10.3, 4.53, -10.89)
- `String`  (double and single quoted characters, numericals and even space)
- `boolean` (either True or False, *the first letter has to be capital*) 

#### To determine the datatype of the value, use the `type` function

```python
name = "this is new name"
print (type(name))
```

```python
a = 1
print (type(a))
```

#### Typecasting - method of converting one datatype to another
-  converting `int` to `float` will add additional decimal point
-  converting a `float` to a `int` will print only the whole number
-  converting a `int` to a `bool` 
    - `1` will converted to **`True`**
    - `0` will be converted to **`False`**
- converting a `int` to `str` will directly convert integer value to a string


```python
name = 1
first_value = str(bool(name))
second_value = "value two"
print(first_value + " " + second_value)
```

### Slicing Operation in a String  
Syntax : **string`[start:end+1:step]`** <br>
-  `start` is starting point of the string
- `end` is one position next to the desired ending position
- `step` is how much position to be skipped while slicing


```python
newInputString = "I love my India"
slicedString1 = newInputString[10:]
FirstSetString2 = newInputString[:7]
print(slicedString1)
print(FirstSetString2)
```

### Membership Operation in a String
Syntax : 
- 'value' **`in`** string. if value is in String, return True, else false
- 'value' **`not in`** string. if value is in String, return True, else false

```python
membershipString = "Greendzine is a premier in Indian electric vehicle industry"
print("Greendzine" in membershipString)
print("batterySwapping" in membershipString)
print("Make in India" not in membershipString)
```

- **format()**   - It is type like a choose the option to fill in the blank.

```python
value_0 = 11
value_1 = "Srinivas"
value_2 = "Varadhan"
print("My name is {0} {1} and my employee id is {2}".format(value_1, value_2, value_0))
```

## String Exercise Problem

##### Problem Statement
In a certain encrypted message which has information about the location(area, city), the characters are jumbled such that first character of the first word is followed by the first character of the second word, then it is followed by second character of the first word and so on

In other words, let’s say the location is bandra,mumbai

The encrypted message says ‘bmaunmdbraai’.

**Sample Input:**
bmaunmdbraai

**Sample Output:**
bandra,mumbai

Let’s say the size or length of the two words wouldn’t match then the smaller word is appended with # and then encrypted in the above format.

```python
input_str = input()
raw_msg_1 = input_str[::2]
raw_msg_2 = input_str[1::2]
message1 = raw_msg_1.strip('#')
message2 = raw_msg_2.strip('#')
print(message1 + "," + message2)
```
