```python
# Read the dataframe with Date_Time variable in the correct form and as Index
weather = pd.read_csv(file_loc, parse_dates=True, index_col = 'Date_Time')

# Check the created dataframe
weather.head()
```

```python
# Delete the index label
weather.index.name
```

```python
# Fetch the entries for the date: 'February 5, 2010'
weather.loc['February 5, 2010']
```

```python
# Slice all the entries between 9 AM and 10 PM on 11th October
weather.loc['October 11, 2010 9:00:00' : 'October 11, 2010 22:00:00']
```

```python
# Slice all the entries between 15th and 31st October
oct_second_half = weather.loc['October 15, 2010' : 'October 31, 2010']
oct_second_half.index
```

```python
# Plot the trend of the data for the entries obtained above
plt.plot(oct_second_half.index, oct_second_half['Temperature'])
plt.plot(oct_second_half.index,oct_second_half['DewPoint'])
plt.plot(oct_second_half.index,oct_second_half['Pressure'])
plt.show()

```
